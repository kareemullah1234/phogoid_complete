```python
import numpy 
from matplotlib import pyplot
%matplotlib inline
```

```python
# Create the time grid.
T = 100.0  # length of the time-interval
dt = 0.1  # time-step size
N = int(T / dt) + 1  # number of time steps
#t = numpy.arange(0.0,100+dt,0.02)  # time grid
g = 9.81
t = numpy.linspace(0.0, T, num=N)
print(t)
print(t.shape)
#print(z)
#print(z.shape)
```

```python
z0 = 100  # give initial values  
b0 = 10 #gust velocity
zt =100
g=9.81
u = numpy.array([z0, b0])
print(u)

u.shape
```

```python
z = numpy.zeros(N) #We both created and initialized with zeros everywhere the solution vector z
print(z.shape)
print(z) 
z[0] =z0
#In the last line above, we assign the initial value to the first element of the solution vector: z[0].
```

```python

#u = numpy.array([z0{u0}, b0{u1}])
for n in range(1,N):
    rhs =  numpy.array([u[1],g*(1- u[0]/zt)])
    u = u + dt * rhs
    z[n] = u[0]
print('z=',z)
print('n=',n)
print('u=',u)


```

```python
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size']   = 16
```

```python
pyplot.figure(figsize=(9.0,4.0))
pyplot.title('elevation of phugoid over time')
pyplot.xlabel('Time[s]')
pyplot.ylabel('Elevation[m]')
pyplot.xlim(t[0],t[-1])
pyplot.ylim(0,200)
pyplot.grid()
pyplot.plot(t,z,color='C0',linestyle='--',linewidth=2)
```

```python
z_exact = (b0 * (zt / g)**0.5 * numpy.sin((g / zt)**0.5 * t) +
           (z0 - zt) * numpy.cos((g / zt)**0.5 * t) + zt)

z_exact
```

```python
# Plot the numerical solution and the exact solution.
pyplot.figure(figsize=(9.0, 4.0))  # set the size of the figure
pyplot.title('Elevation of the phugoid over the time')  # set the title
pyplot.xlabel('Time [s]')  # set the x-axis label
pyplot.ylabel('Elevation [m]')  # set the y-axis label
pyplot.xlim(t[0], t[-1])  # set the x-axis limits
pyplot.ylim(40.0, 160.0)  # set the y-axis limits
pyplot.grid()  # set a background grid to improve readability
pyplot.plot(t, z, label='Numerical',
            color='C0', linestyle='-', linewidth=2)
pyplot.plot(t, z_exact, label='Analytical',
            color='C1', linestyle='-', linewidth=2)
pyplot.legend();  # set the legend
```

```python
dt_values = [0.1,0.05,0.01,0.0050, 0.001,0.0001]
T=100
z_values = []
t1 = []
for dt in dt_values:
    N = int(T/dt) + 1  #number of time steps
    
    print(N)
    
    numpy.array([N])
    #print(N)
    t =numpy.linspace(0.0,T,num=N) # time grid
    t1.append(t)
    print(t)
    # intitial conditions are
    u = numpy.array([z0,b0])
    z = numpy.empty_like(t)
    z[0] = z0
    
    for n in range(1,N):
        rhs = numpy.array([u[1],g*(1-u[0]/zt)])
        u = u + dt * rhs
        z[n] = u[0] # storing elevation at next step n+1
    z_values.append(z)
print(z_values) 

print(z.shape)
```

```python
def l1_error(z,z_exact,dt): 
    error = dt * numpy.sum(numpy.abs(z-z_exact))  
    return error                       

#z[n] = u[0]
 #z_exact = (b0 * (zt / g)**0.5 * numpy.sin((g / zt)**0.5 * t) +
#(z0 - zt) * numpy.cos((g / zt)**0.5 * t) + zt)
                            
```

```python
error_values =[]

for z,dt in zip(z_values,dt_values):  # z_values = [] for dt_values = [0.1,0.05,0.01,0.0050, 0.001,0.0001]
    N= int(T/dt) + 1            #T =100, dt=0.1,0.05,0.01,0.0050, 0.001,0.0001
    print(N)
    t =numpy.linspace(0.0,T,num = N)
    print(t.shape)
    z_exact = (b0 * (zt / g)**0.5 * numpy.sin((g / zt)**0.5 * t) +
               (z0 - zt) * numpy.cos((g / zt)**0.5 * t) + zt)
    print(z_exact)
    print(z_exact.shape)
    error_values.append(l1_error(z,z_exact,dt))
print(error_values)
```

```python
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size']   = 200
pyplot.figure(figsize=(100,100))
pyplot.title('error vs time step size')
pyplot.xlabel('time step size for each calculation')
pyplot.ylabel('sum of error difference')
pyplot.grid()
pyplot.plot(dt_values,error_values,color='C0',linestyle='-',marker='o',linewidth='3')
#pyplo
```
